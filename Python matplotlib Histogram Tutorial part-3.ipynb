# ploting Histogram

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
import random
```

```python
'''
for histogram banane ke liye random data generate karne ke liye random module and
random module ko array me convert karne ke liye numpy ka use kiya hai'''
#for student age random data 
```

```python
ml_student_age=np.random.randint(18,40,(100))
py_student_age=np.random.randint(15,40,(100))
print(ml_student_age)
print(py_student_age)
```

```python
plt.hist(ml_student_age)
plt.show()
```

```python
plt.hist(ml_student_age)
plt.title('machine learning student and age',fontsize=15)
plt.show()
```

```python
plt.hist(ml_student_age)
plt.title('machine learning student and age',fontsize=15)
plt.xlabel('student age category',fontsize=12)
plt.ylabel('no. of student',fontsize=13)
plt.show()
```

```python
plt.hist(ml_student_age,bins=[15,20,25,30,35,40])
plt.title('machine learning student and age',fontsize=15)
plt.xlabel('student age category',fontsize=12)
plt.ylabel('no. of student',fontsize=13)
plt.show()
```

```python
#rwith=0.8 parameter
plt.hist(ml_student_age,bins=[15,20,25,30,35,40],rwidth=0.8)
plt.title('machine learning student and age',fontsize=15)
plt.xlabel('student age category',fontsize=12)
plt.ylabel('no. of student',fontsize=13)
plt.show()
```

```python
#histtype='bar'
plt.hist(ml_student_age,bins=[15,20,25,30,35,40],rwidth=0.8,histtype='step')
plt.title('machine learning student and age',fontsize=15)
plt.xlabel('student age category',fontsize=12)
plt.ylabel('no. of student',fontsize=13)
plt.show()
```

```python
plt.hist(ml_student_age,bins=[15,20,25,30,35,40],rwidth=0.8,histtype='bar')
plt.title('machine learning student and age',fontsize=15)
plt.xlabel('student age category',fontsize=12)
plt.ylabel('no. of student',fontsize=13)
plt.show()
```

```python
#align='mid'
plt.hist(ml_student_age,bins=[15,20,25,30,35,40],rwidth=0.8,histtype='bar',align='mid')
plt.title('machine learning student and age',fontsize=15)
plt.xlabel('student age category',fontsize=12)
plt.ylabel('no. of student',fontsize=13)
plt.show()
```

```python
#orientation='vertical
plt.hist(ml_student_age,bins=[15,20,25,30,35,40],rwidth=0.8,histtype='bar',align='mid',orientation='horizontal')
plt.title('machine learning student and age',fontsize=15)
plt.xlabel('student age category',fontsize=12)
plt.ylabel('no. of student',fontsize=13)
plt.show()
```

```python
#orientation='vertical
plt.hist(ml_student_age,bins=[15,20,25,30,35,40],rwidth=0.8,histtype='bar',align='mid',orientation='vertical')
plt.title('machine learning student and age',fontsize=15)
plt.xlabel('student age category',fontsize=12)
plt.ylabel('no. of student',fontsize=13)
plt.show()
```

```python
#color
plt.hist(ml_student_age,bins=[15,20,25,30,35,40],rwidth=0.8,histtype='bar',align='mid',orientation='vertical',color='g')
plt.title('machine learning student and age',fontsize=15)
plt.xlabel('student age category',fontsize=12)
plt.ylabel('no. of student',fontsize=13)
plt.show()
```

```python
#label
plt.hist(ml_student_age,bins=[15,20,25,30,35,40],rwidth=0.8,histtype='bar',align='mid',orientation='vertical',color='g',
         label='ml_student')
plt.title('machine learning student and age',fontsize=15)
plt.xlabel('student age category',fontsize=12)
plt.ylabel('no. of student',fontsize=13)
plt.legend(loc=2)
plt.show()
```

```python
#for 2 student information in 1 histogram
plt.hist(ml_student_age,bins=[15,20,25,30,35,40],rwidth=0.8,histtype='bar',align='mid',orientation='vertical',color='g',
         label='ml_student')
plt.hist(py_student_age,bins=[15,20,25,30,35,40],rwidth=0.8,histtype='bar',align='mid',orientation='horizontal',color='y',
         label='py_student')
plt.title('machine learning student and age',fontsize=15)
plt.xlabel('student age category',fontsize=12)
plt.ylabel('no. of student',fontsize=13)
plt.legend(loc=2)
plt.show()
```

```python
plt.hist([ml_student_age,py_student_age],bins=[15,20,25,30,35,40],rwidth=0.8,histtype='bar',align='mid',orientation='vertical',color=['g','y'],
         label='ml_student')
#plt.hist(py_student_age,bins=[15,20,25,30,35,40],rwidth=0.8,histtype='bar',align='mid',orientation='horizontal',color='y',
   #      label='py_student')
plt.title('machine learning student and age',fontsize=15)
plt.xlabel('student age category',fontsize=12)
plt.ylabel('no. of student',fontsize=13)
plt.legend(loc=2)
plt.show()
```

```python
#histogram bda karne ke liye
#plt.figure(figsize=(sizeno.,sizeno.))function ka use karte hai
plt.figure(figsize=(15,9))
plt.hist([ml_student_age,py_student_age],bins=[15,20,25,30,35,40],rwidth=0.8,histtype='bar',align='mid',orientation='vertical',color=['g','y'],
         label=['ml_student','py_student'])
#plt.hist(py_student_age,bins=[15,20,25,30,35,40],rwidth=0.8,histtype='bar',align='mid',orientation='horizontal',color='y',
   #      label='py_student')
plt.title('machine learning student and age',fontsize=15)
plt.xlabel('student age category',fontsize=12)
plt.ylabel('no. of student',fontsize=13)
plt.legend(loc=2)
#plt.grid(color='w',linestyle='-',linewidth=1)
plt.show()
```
