# ploting bar chart

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
#data set lete hai for exaple class data set
#classes=['python','r','artificial intrlligence','machine learning','data sciemce']
classes=['python','R','AI','ML','DS']
class1_student=[30,10,20,25,10]#out of 100 student in each class 100 me se class me kitne log hai
class2_student=[40,5,20,20,10]
class3_student=[35,5,30,15,15]
```

```python
#plt.bar(x,y)mahtod
plt.bar(classes,class1_student)
```

```python
plt.barh(classes,class1_student)
plt.show()
```

```python
#width=size no. of width
plt.bar(classes,class1_student,width=0.3)
```

```python
#align='edge'
plt.bar(classes,class1_student,width=0.3,align='edge')
```

```python
#color='g'
plt.bar(classes,class1_student,width=0.3,align='center',color='g')
plt.show()
```

```python
#for bar edge color edgecolor='colorname'
plt.bar(classes,class1_student,width=0.3,align='center',color='g',edgecolor='m')
plt.show()
```

```python
#agr bar ke line width ko bdana ho to= linewidth=no.of width size
plt.bar(classes,class1_student,width=0.3,align='center',color='g',edgecolor='m',linewidth=5)
plt.show()
```

```python
#if bar ka color light karna ho to alpha='value between o to 1',alpha=0.4
plt.bar(classes,class1_student,width=0.4,color='g',align='center',edgecolor='m',linewidth=4,alpha=0.6)
plt.show()
```

```python
#agar bar ke line style ko change karna ho to =linestyle="--"
plt.bar(classes,class1_student,width=0.4,color='g',align='center',edgecolor='m',linewidth=4,alpha=0.6,linestyle="--")
plt.show()
```

```python
plt.bar(classes,class1_student,width=0.4,color='g',align='center',edgecolor='m',linewidth=4,alpha=0.6,linestyle="--",label='class1_student')
plt.title('classes and student',fontsize=15)
plt.legend()
plt.xlabel('classses',fontsize=10)
plt.ylabel('class_1student',fontsize=10)
plt.show()
```

```python
#if we want hide our graph visual=false
plt.bar(classes,class1_student,width=0.4,color='g',align='center',edgecolor='m',linewidth=4,
        alpha=0.6,linestyle="--",label='class1_student',visible=True)
plt.title('classes and student',fontsize=15)
plt.legend()
plt.xlabel('classses',fontsize=10)
plt.ylabel('class_1student',fontsize=10)
plt.show()
```
