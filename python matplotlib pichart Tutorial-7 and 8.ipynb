# pie plot

```python
import matplotlib.pyplot as plt
```

```python
#manual data set lete hai
classes=['python','R','Machine learning','Artificial intelligence','Data science']
class1_student=[45,15,35,25,30]
```

```python
#pie plot
plt.pie([1])
plt.show()
```

```python
plt.pie(class1_student)
plt.show()
```

```python
#manual data set lete hai
classes=['python','R','Machine learning','Artificial intelligence','Data science']
class1_student=[45,15,35,25,30]
plt.pie(class1_student,labels=classes)
plt.show()
```

```python
#machie learning slicing
classes=['python','R','Machine learning','Artificial intelligence','Data science']
class1_student=[45,15,35,25,30]
explode=[0,0,0.1,0,0]
plt.pie(class1_student,labels=classes,explode=explode)
plt.show()
```

```python
#agar python ko bhi slice karna hai to
classes=['python','R','Machine learning','Artificial intelligence','Data science']
class1_student=[45,15,35,25,30]
explode=[0.1,0,0.1,0,0]
plt.pie(class1_student,labels=classes,explode=explode)
plt.show()
```

```python
#direct slicing in parameter
classes=['python','R','Machine learning','Artificial intelligence','Data science']
class1_student=[45,15,35,25,30]
plt.pie(class1_student,labels=classes,explode=[0.1,0,0.1,0,0])
plt.show()
```

```python
#direct slicing se data science slice
classes=['python','R','Machine learning','Artificial intelligence','Data science']
class1_student=[45,15,35,25,30]
plt.pie(class1_student,labels=classes,explode=[0.1,0,0.1,0,0.1])
plt.show()
```

```python
#color parameter
classes=['python','R','Machine learning','Artificial intelligence','Data science']
class1_student=[45,15,35,25,30]
colors=['g','b','k','y','m']
plt.pie(class1_student,labels=classes,explode=[0.1,0,0.1,0,0.1],colors=colors)
plt.show()
```

```python
#data percentage for -autopct=%0.1%%
classes=['python','R','Machine learning','Artificial intelligence','Data science']
class1_student=[45,15,35,25,30]
colors=['g','b','k','y','m']
plt.pie(class1_student,labels=classes,explode=[0.05,0,0.02,0,0.08],colors=colors,autopct='%0.1f%%')
plt.show()
```

```python
#shadow 
classes=['python','R','Machine learning','Artificial intelligence','Data science']
class1_student=[45,15,35,25,30]
colors=['g','b','k','y','m']
plt.pie(class1_student,labels=classes,explode=[0.05,0,0.02,0,0.08],colors=colors,autopct='%0.1f%%',shadow=True)
plt.show()
```

```python
#radius
classes=['python','R','Machine learning','Artificial intelligence','Data science']
class1_student=[45,15,35,25,30]
colors=['g','b','k','y','m']
plt.pie(class1_student,labels=classes,explode=[0.05,0,0.02,0,0.08],colors=colors,autopct='%0.1f%%',shadow=True,radius=2)
plt.show()
```

```python
#startangle
classes=['python','R','Machine learning','Artificial intelligence','Data science']
class1_student=[45,15,35,25,30]
colors=['g','b','k','y','m']
plt.pie(class1_student,labels=classes,explode=[0.05,0,0.02,0,0.08],colors=colors,autopct='%0.1f%%',
        shadow=True,radius=2,startangle=90)
plt.show()
```

```python
#textprops
classes=['python','R','Machine learning','Artificial intelligence','Data science']
class1_student=[45,15,35,25,30]
colors=['g','b','m','y','r']
textprops={'fontsize':16}
plt.pie(class1_student,labels=classes,explode=[0.05,0,0.02,0,0.08],colors=colors,autopct='%0.1f%%',
        shadow=True,radius=2,startangle=90,textprops=textprops)
plt.show()
```

```python

classes=['python','R','Machine learning','Artificial intelligence','Data science']

class1_student=[45,15,35,25,30]
colors=['g','b','m','y','r']
textprops={'fontsize':16}

wedgeprops={'linewidth':4,'width':2,'edgecolor':'y'}
           
plt.pie(class1_student,
        labels=classes,
        explode=[0.05,0,0.02,0,0.08],
        colors=colors,autopct='%0.1f%%',
        shadow=True,
        radius=1.1,
        startangle=90,
        textprops=textprops,
        labeldistance=1.4,
        pctdistance=0.6,
        counterclock=True,
        wedgeprops=wedgeprops,
        center=(2,3),
        frame=True,
        rotatelabels=True
       )
           
           
#plt.figure(figsize=(8,4)
plt.show()
           
```

```python
classes=['python','R','Machine learning','Artificial intelligence','Data science']
class1_student=[45,15,35,25,30]
colors=['g','b','m','y','r']
textprops={'fontsize':16}
plt.pie(class1_student,labels=classes,explode=[0.05,0,0.02,0,0.08],colors=colors,autopct='%0.1f%%',
        shadow=True,radius=2,startangle=90,textprops=textprops)
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')
classes=['python','R','Machine learning','Artificial intelligence','Data science']
class1_student=[45,15,35,25,30]
colors=['g','b','m','y','r']
textprops={'fontsize':16}
plt.pie(class1_student,labels=classes,explode=[0.05,0,0.02,0,0.08],colors=colors,autopct='%0.1f%%',
        shadow=True,radius=2,startangle=90,textprops=textprops)
plt.legend(loc=2)
plt.show()
```
