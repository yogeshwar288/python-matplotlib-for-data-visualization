```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
days=np.arange(1,16)
days
```

```python
print(days)
```

```python
betul_temp=[26.55,27.88,25.22,26.30,20.55,25.44,27.22,30.22,31.22,35.22,25.22,27.22,30.25,28.22,25.22]
```

```python
plt.plot(days,betul_temp)
plt.show()
```

```python
# for graph title
plt.plot(days,betul_temp)
plt.title('betul_temprature')
plt.show()
```

```python
#for axis name
```

```python
plt.plot(days,betul_temp)
plt.title('betul temprature')
plt.xlabel('days')
plt.ylabel('betul_temprature')
plt.show()
```

```python
#but if we want x,y axis starting point 0 than
plt.plot(days,betul_temp)
plt.axis([0,20,0,45])
plt.title('betul temprature')
plt.xlabel('days')
plt.xlabel('temprature')
plt.show()
```
