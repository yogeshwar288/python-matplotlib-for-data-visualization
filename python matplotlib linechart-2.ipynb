```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
days=np.arange(1,16)
days
```

```python
betul_temp=[26.55,27.88,25.22,26.30,20.55,25.44,27.22,30.22,31.22,35.22,25.22,27.22,30.25,28.22,25.22]
```

```python
plt.plot(days,betul_temp)
```

```python
plt.plot(days,betul_temp)
plt.title('betul temprature')
plt.xlabel('days')
plt.ylabel('temprature')
plt.axis([0,25,0,45])
plt.show()
```

# line chat 



# plot() method parameter

```python
#for change our curve line color
plt.plot(days,betul_temp,color='g')
plt.title('betul temprature')
plt.xlabel('days')
plt.ylabel('temprature')
plt.axis([0,25,0,45])
plt.show()
```

```python
#if we want temprature data data point  doted 
plt.plot(days,betul_temp,color='g',marker='o')
plt.title('betul temprature')
plt.xlabel('days')
plt.ylabel('temprature')
plt.axis([0,25,0,45])
plt.show()
```

```python
#we have marker 4-5 type in matplotlib
#'.', ',' 'o', '>' , '<', '1'
```

```python
#if we wanted doted-doted curve line

plt.plot(days,betul_temp,color='g',marker='o',linestyle='--')
plt.title('betul temprature')
plt.xlabel('days')
plt.ylabel('temprature')
plt.axis([0,25,0,45])
plt.show()
```

```python
plt.plot(days,betul_temp,color='g',marker='o',linestyle='--')
plt.title('betul temprature')
plt.xlabel('days')
plt.ylabel('temprature')
plt.axis([0,20,0,45])
plt.show()
```

```python
#if we want to chnge  line width
plt.plot(days,betul_temp,color='g',marker='o',linestyle='--',linewidth=3)
plt.title('betul temprature')
plt.xlabel('days')
plt.ylabel('temprature')
plt.axis([0,20,0,45])
plt.show()
```

```python
#if we want to chnge temprature data data point  doted marker size
plt.plot(days,betul_temp,color='g',marker='o',linestyle='--',linewidth=3, markersize=8)
plt.title('betul temprature')
plt.xlabel('days')
plt.ylabel('temprature')
plt.axis([0,20,0,45])
plt.show()
```

```python
#sort method for using color='g',marker='o',linestyle='--' in same time
#we can use directly 'go--' 
plt.plot(days,betul_temp,'go--',linewidth=3, markersize=8)
plt.title('betul temprature')
plt.xlabel('days')
plt.ylabel('temprature')
plt.axis([0,20,0,45])
plt.show()
```

# font size

```python
#font size tital parameter yA LABLE PARAMETER ME DE SAKTE HAI JIS BHI STRING KI FONT SIZE BDANE KI AAVSYAKTA HOTI HAI
plt.plot(days,betul_temp,'go--',linewidth=3, markersize=8)
plt.title('betul temprature',fontsize=20)
plt.xlabel('days')
plt.ylabel('temprature')
plt.axis([0,20,0,45])
plt.show()
```

```python
plt.plot(days,betul_temp,'bo--',linewidth=3, markersize=8)
plt.title('betul temprature',fontsize=15)
plt.xlabel('days',fontsize=12)
plt.ylabel('temprature',fontsize='12')
plt.axis([0,20,0,45])
plt.show()
```

# legend function
curve line ko lable dene ke liye use hota hai plt.legend(['string ke rup me line lable likhte hai'])

```python
plt.plot(days,betul_temp,'go--',linewidth=3, markersize=8)
plt.title('betul temprature',fontsize=15)
plt.xlabel('days',fontsize=12)
plt.ylabel('temprature',fontsize='12')
plt.axis([0,20,0,45])
plt.legend(['temp_line'])
plt.show()
```

```python
#legand([],loc=1), curveline label  ki location chnge karne ke liye
plt.plot(days,betul_temp,'go--',linewidth=3, markersize=8)
plt.title('betul temprature',fontsize=15)
plt.xlabel('days',fontsize=12)
plt.ylabel('temprature',fontsize='12')
plt.axis([0,20,0,45])
plt.legend(['temp_line'],loc=4)
plt.show()
```

```python
#lable parametar same work karta hai legand  ki tarah ise plot me use karte hai
plt.plot(days,betul_temp,'go--',linewidth=3, markersize=8,label='Templine')
plt.title('betul temprature',fontsize=15)
plt.xlabel('days',fontsize=12)
plt.ylabel('temprature',fontsize='12')
plt.axis([0,20,0,45])
plt.legend(loc=4)
plt.show()
```
