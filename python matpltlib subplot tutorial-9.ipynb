```python
#figure function
plt.figure(figsize=(16,9))
plt.subplot(321)

###line chart

days=np.arange(1,16)
betul_temp=[26.55,27.88,25.22,26.30,20.55,25.44,27.22,30.22,31.22,35.22,25.22,27.22,30.25,28.22,25.22]
bhopal_temp=[30.2,32.2,38.2,28,25,28,30,25,32,28,25,28,29,30,28.52]
indore_temp=[20,25,26,24,25,21,30,31,28,21,20,18,20,22,24]
#plt.figure(figsize=(16,9))
style.use('ggplot')
plt.plot(days,betul_temp,'go--',linewidth=3, markersize=8,label='betul Templine')
plt.plot(days,bhopal_temp,'bo--',linewidth=2,markersize=10,label='bhopal teprature')
plt.plot(days,indore_temp,'ko--',linewidth=2,markersize=10,label='indore teprature')
plt.title(' temprature',fontsize=15)
plt.xlabel('days',fontsize=12)
plt.ylabel('temprature',fontsize='12')
plt.axis([0,20,0,45])
plt.legend(loc=4)
plt.grid(color='w',linestyle='-',linewidth=1)

#####histogram chart

plt.subplot(3,2,2)
style.use('ggplot')
ml_student_age=np.random.randint(18,40,(100))
py_student_age=np.random.randint(15,40,(100))
#plt.figure(figsize=(15,9))
plt.hist([ml_student_age,py_student_age],bins=[15,20,25,30,35,40],rwidth=0.8,histtype='bar',align='mid',orientation='vertical',color=['g','y'],
         label=['ml_student','py_student'])
#plt.hist(py_student_age,bins=[15,20,25,30,35,40],rwidth=0.8,histtype='bar',align='mid',orientation='horizontal',color='y',
   #      label='py_student')
plt.title('machine learning student and age',fontsize=15)
plt.xlabel('student age category',fontsize=12)
plt.ylabel('no. of student',fontsize=13)
plt.legend(loc=2)
#plt.grid(color='w',linestyle='-',linewidth=1)
#plt.show()

######bar chart

plt.subplot(3,2,3)

#data set lete hai for exaple class data set
#classes=['python','r','artificial intrlligence','machine learning','data sciemce']
classes=['python','R','AI','ML','DS']
class1_student=[30,10,20,25,10]#out of 100 student in each class 100 me se class me kitne log hai
class2_student=[40,5,20,20,10]
class3_student=[35,5,30,15,15]
#plt.figure(figsize=(15,8))
classes_index=np.arange(len(classes))
width=0.2
plt.bar(classes_index,class1_student,width,color='g',align='center',edgecolor='b',linewidth=4,
        alpha=0.6,label='class1_student',visible=True)
plt.bar(classes_index+width,class2_student,width,color='b',align='center',edgecolor='k',linewidth=4,
        alpha=0.6,label='class2_student',visible=True)
plt.bar(classes_index+width+width,class2_student,width,color='y',align='center',edgecolor='k',linewidth=4,
        alpha=0.6,label='class3_student',visible=True)
plt.xticks(classes_index+width,classes)
#plt.bar(classes,class3_student,width=0.4,color='y',align='center',edgecolor='g',linewidth=4,
      # alpha=0.6,label='class3_student',visible=True)
style.use('ggplot')
plt.title('classes and student',fontsize=15)
plt.legend()
plt.xlabel('classses',fontsize=10)
plt.ylabel('class_1student',fontsize=10)

###scatter chart

plt.subplot(3,2,4)


df_google_playstore_data=pd.read_csv('G:\\csv\\googleplaystore.csv',nrows=4000)
df_google_playstore_data
x=df_google_playstore_data['Rating']
y=df_google_playstore_data['Reviews']
z=df_google_playstore_data['Installs']
style.use('ggplot')
#plt.figure(figsize=(16,9))
plt.scatter(x,y,color='g',edgecolor='y',alpha=0.6,s=100,linewidth=4,marker="*",label='reviews')
plt.scatter(x,z,color='y',edgecolor='m',alpha=0.6,s=30,linewidth=3,marker="o",label='install')
plt.scatter(x,df_google_playstore_data['Price'],color='w',edgecolor='g',s=20,linewidth=2,marker=".",alpha=0.4,label='price')
plt.legend()
plt.title('googleplay store rating and reviews',fontsize=20)
plt.xlabel('ratings',fontsize=12)
plt.ylabel('reviews,install and prices',fontsize=12)



###pi chart
plt.subplot(3,2,5)

classes=['python','R','Machine learning','Artificial intelligence','Data science']
class1_student=[45,15,35,25,30]
colors=['g','b','k','y','m']
plt.pie(class1_student,labels=classes,explode=[0.05,0,0.02,0,0.08],colors=colors,autopct='%0.1f%%',shadow=True)
#plt.show()

plt.subplot(3,2,6,projection='polar',facecolor='k',frameon=False)
plt.show()
```
